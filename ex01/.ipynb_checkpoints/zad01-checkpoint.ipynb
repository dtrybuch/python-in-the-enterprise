```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression

boston_data = load_boston()
df = pd.DataFrame(data=boston_data.data, columns=boston_data.feature_names)
print(df.head())
```

```python
df1 = df.iloc[:,[0,9]].copy()
df1.hist(bins = 20)
```

```python
X_CRIM = df['CRIM']
Y_target = boston_data.target
X_CRIM = np.array(X_CRIM).reshape(-1,1)
Y_target = np.array(Y_target).reshape(-1,1)
reg = LinearRegression().fit(X_CRIM,Y_target)
plt.scatter(x=X_CRIM,y=Y_target)
plt.plot(X_CRIM,reg.predict(X_CRIM),  color='red')
plt.ylabel('target (price)')
plt.xlabel('CRIM')
plt.show()
```

```python
X_TAX = df['TAX']
X_TAX = np.array(X_TAX).reshape(-1,1)
reg = LinearRegression().fit(X_TAX,Y_target)
plt.scatter(x=X_TAX,y=Y_target)
plt.plot(X_TAX,reg.predict(X_TAX), color='red')
plt.ylabel('target (price)')
plt.xlabel('TAX')
plt.show()
```

When I look at the first plot with crime rate per capita I think that the price is not so much linear dependant of it. The price is fitted to negative values so it is not appropriate approximation. I would suggest more hyperbolic function to use.

However when we look at the second plot with full-value property-tax rate per $10,000 we can see that the data is more random. We can fit it with linear function but it is weakly dependant of linear function.  I cannot say which function should be more dependant because the data is too much scattered.
